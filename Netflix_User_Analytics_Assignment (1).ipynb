```python
import pandas as pd

df = pd.read_csv("Dataset 2 (1).csv")

df.head()
```

```python
print(df.isnull().sum())
```

```python
print("Average Age =", df["Age"].mean())

print("Average Watch Hours =", df["WatchHoursPerWeek"].mean())

print("Average Monthly Spend =", df["MonthlySpend"].mean())

print("\nSubscription Counts")
print(df["SubscriptionType"].value_counts())

renewal_percentage = (
    (df["SubscriptionRenewed"] == "Yes").sum()
    / len(df)
) * 100

print("\nRenewal Percentage =", renewal_percentage)
```

```python
from sklearn.preprocessing import LabelEncoder

df_ml = df.copy()

le_gender = LabelEncoder()
le_sub = LabelEncoder()
le_genre = LabelEncoder()
le_renew = LabelEncoder()

df_ml["Gender"] = le_gender.fit_transform(df_ml["Gender"])
df_ml["SubscriptionType"] = le_sub.fit_transform(df_ml["SubscriptionType"])
df_ml["FavoriteGenre"] = le_genre.fit_transform(df_ml["FavoriteGenre"])
df_ml["SubscriptionRenewed"] = le_renew.fit_transform(df_ml["SubscriptionRenewed"])

print(df_ml.head())
```

```python
from sklearn.model_selection import train_test_split

X = df_ml.drop(["UserID", "SubscriptionRenewed"], axis=1)

y = df_ml["SubscriptionRenewed"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

dt_accuracy = accuracy_score(y_test, y_pred_dt)

print("Decision Tree Accuracy =", dt_accuracy)

cm = confusion_matrix(y_test, y_pred_dt)

print("\nConfusion Matrix")
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

knn_accuracy = accuracy_score(y_test, y_pred_knn)

print("KNN Accuracy =", knn_accuracy)

print("\nComparison")
print("Decision Tree =", dt_accuracy)
print("KNN =", knn_accuracy)
```

```python
from sklearn.linear_model import LinearRegression

X_reg = df_ml.drop(["UserID", "MonthlySpend"], axis=1)

y_reg = df_ml["MonthlySpend"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

lr = LinearRegression()

lr.fit(X_train_reg, y_train_reg)

print("Linear Regression Model Trained Successfully")
```

Based on the analysis, factors such as MonthlySpend, WatchHoursPerWeek, SubscriptionType, AdClicks, and Age appear to have the greatest influence on subscription renewal. Users who spend more time watching content and invest more money in their subscriptions are generally more likely to renew.

**2. Why is subscription renewal a classification problem?**

Subscription renewal is a classification problem because the target variable has two possible categories: Yes (renewed) and No (not renewed). The objective is to classify users into one of these categories.

**3. Why is monthly spending a regression problem?**

Monthly spending is a regression problem because it is a continuous numerical value. The goal is to predict a specific amount rather than assign a category.

**4. Which algorithm performed better for renewal prediction?**

The K-Nearest Neighbors (KNN) algorithm performed better than the Decision Tree model. The Decision Tree achieved an accuracy of 56%, while KNN achieved an accuracy of 63.33%. Therefore, KNN provided better prediction performance for subscription renewal.
