## Q4. Why is KNN considered a supervised learning algorithm? Can KNN perform classification if labels are not available?

KNN is a supervised learning algorithm because it uses training data that already contains labels.

No, KNN cannot perform classification if labels are not available because it needs labeled neighbors to determine the class of a new data point.

## Q5. Suppose a KNN model uses K = 5 and the nearest neighbors are:

Yes, Yes, No, Yes, No

The prediction can be verified using the following code:

```python
neighbors = ["Yes", "Yes", "No", "Yes", "No"]

yes_count = neighbors.count("Yes")
no_count = neighbors.count("No")

if yes_count > no_count:
    print("Prediction: Yes")
else:
    print("Prediction: No")
```

**Output:** Prediction: Yes

The prediction is Yes because it appears 3 times out of 5 neighbors and KNN uses majority voting.

## Q9. What is the purpose of a Confusion Matrix? How is it different from a Correlation Matrix?

### Confusion Matrix

A Confusion Matrix is used to evaluate the performance of a classification model.

It contains:
- True Positive (TP)
- True Negative (TN)
- False Positive (FP)
- False Negative (FN)

### Correlation Matrix

A Correlation Matrix shows the relationship between variables.

### Difference

- Confusion Matrix evaluates model performance.
- Correlation Matrix measures relationships between variables.
